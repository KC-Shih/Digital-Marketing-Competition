##     1)  SVM for C_Other
### 1. 只用 training dataset 來訓練資料：不用分割資料集

```python
import pandas as pd
import time
import numpy as np
from sklearn import svm
import time
from sklearn import metrics
from sklearn.externals import joblib
```

```python
df = pd.read_csv(r"D:\1091\行銷數據分析\新資料集\C_Other_train.csv")
df = df.drop(labels=['ssid','play_mode'],axis='columns')  # 去除部分欄位
df
```

```python
x = [i for i in df.columns if i not in ["view_50or_click"]] 
X_train = df[x]
y_train = df[["view_50or_click"]]  
```

```python
#SVM model
#Create a svm Classifier

SVM_C = svm.SVC(C=100, kernel="rbf", gamma=0.1)

# fit model by all columns
# Train the model using the training sets
start_time = time.time()
SVM_C.fit(X_train, y_train.values.ravel())

print("fitting took{}seconds".format(time.time()-start_time))
```

```python
df2 = pd.read_csv(r"D:\1091\行銷數據分析\新資料集\C_Other_verify.csv")
df2 = df2.drop(labels=['ssid','play_mode'],axis='columns')  # 去除部分欄位
df2
```

```python
x_v = [i for i in df2.columns if i not in ["view_50or_click"]] 
X_verify = df2[x_v]
y_verify = df2[["view_50or_click"]]  
```

```python
#Predict the response for test dataset
start_time = time.time()
y_pred_SVM_C = SVM_C.predict(X_verify)

print("fitting took{}seconds".format(time.time()-start_time))

print("Accuracy:",metrics.accuracy_score(y_verify, y_pred_SVM_C))

#################################

#  Accuracy: 0.7219235560217662

#################################
```

```python
# 儲存模型
joblib.dump(SVM_C,'SVM_C1229.pkl')  # 儲存模型

# saved_model_SVM_C = joblib.load('SVM_C.pkl') # 重新呼叫fit好的模型
```

```python
crosstab = pd.crosstab(y_verify.values.ravel(),y_pred_SVM_C, rownames=['Ture labels'], colnames=['prediction'])

a = int(crosstab[0][:1])
b = int(crosstab[1][:1])
c = int(crosstab[0][1:])
d = int(crosstab[1][1:])
precision = a/(a+b)
recall = a/(a+c)
accuracy = (a+d)/(a+b+c+d)

print(crosstab)

print('precision:',round(precision,3))
print('recall：',round(recall,3))
print('accuracy：',round(accuracy,3))
```
